## Sample Code to collect dataset of human faces and label them

```python
import cv2
import numpy as np
import os
import glob
import time
print("imported")
```

```python



# An instance of app for each person
#dataset path 

dataset_path = "C:\\MyLearnings\\DeepLearning\\Facial Projects\\Face Recognition\\customDataset"
```

```python
print("Enter the name of the person whose images are to e sampled:  ")
name = input()

#create directory of that name in dataset folder
new_dir = os.path.join(dataset_path,name)
os.mkdir(new_dir)
print(f"Directory created at path : {new_dir}")
```

```python
#First 5 sec continuous image save
# At each press of C Capture and save image to dataset 
"""
ALGORITHM
==========
1. Capture each frame,
2.Detect face
3.Get coord
<5sec, add to directory with name as timestamp after resizing
>5sec if c pressed add afetr resizing , else dont
"""
index = 1
face_cascade = cv2.CascadeClassifier('C:\\MyLearnings\\openCV\\HaarCascades\\haarcascade_frontalface_default.xml') 
cap = cv2.VideoCapture(0)
t1 = time.time()
capture = False
ctr = 1
print("1")
while True:
    _,frame = cap.read()
    if ((ctr <100 ) or (capture == True)) :
        ctr +=1
        # convert to gray scale of each frames 
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)    
        # Tuning the second and third parameter is crucial for accurate detection
        #However, a fine tuned value for oneimage may not best fit for another. This is what makes cascades a little inefficient
        # Detects faces of different sizes in the input image 
        faces = face_cascade.detectMultiScale(gray, 1.3, 5) 
        path = ""
        for (x,y,w,h) in faces: 
            print("detected face")
            # To crop ROI
            #cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,0),2) 
            roi_color = frame[y:y+h, x:x+w] 
            img = cv2.resize(roi_color,(48,48))
            img_name = str(index)
            path = os.path.join(new_dir,img_name)
            path = path + ".jpg"
            index+=1
            print(path)
            cv2.imwrite(path,img)
        cv2.imshow("frame",frame)
        capture = False
    key = cv2.waitKey(1)
    if key == ord('c'):
        capture = True
    elif key == ord('q'):
        break
            
cap.release()
cv2.destroyAllWindows()
```

# Code to collect a Single photo of a person and label

```python
path = "C:\\MyLearnings\\DeepLearning\\Facial Projects\\Face Recognition\\SingleShotDataset\\"
name = ""
```

```python
def get_label():
    global name 
    name = input("Enter Name of the person : ")
    return name
    
```

```python
face_cascade = cv2.CascadeClassifier('C:\\MyLearnings\\openCV\\HaarCascades\\haarcascade_frontalface_default.xml') 
cap = cv2.VideoCapture(0)
capture = False
print("1")
while True:
    _,frame = cap.read()
    # convert to gray scale of each frames 
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)    
    # Tuning the second and third parameter is crucial for accurate detection
    #However, a fine tuned value for oneimage may not best fit for another. This is what makes cascades a little inefficient
    # Detects faces of different sizes in the input image 
    faces = face_cascade.detectMultiScale(gray, 1.3, 5) 
    for (x,y,w,h) in faces: 
        # To crop ROI
       # cv2.rectangle(frame,(x,y),(x+w,y+h),(255,255,0),2) 
        roi_color = frame[y:y+h, x:x+w] 
        img = cv2.resize(roi_color,(48,48))
            
            
    cv2.imshow("frame",frame)
    key = cv2.waitKey(1)
    if key == ord('c'):
        img_name = get_label()
        img_name = img_name + ".jpg"
        new_path = os.path.join(path,img_name)
        print(new_path)
        cv2.imshow("img",img)
        cv2.imwrite(new_path,frame)
    elif key == ord('q'):
        break
            
cap.release()
cv2.destroyAllWindows()
```
